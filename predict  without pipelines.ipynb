```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),
                                               df['Survived'],
                                               test_size=0.2,
                                               random_state=42)
```

simple imputer applying

```python
si_age=SimpleImputer() #simple imputer object is created to deal with th emissing values
si_embarked=SimpleImputer(strategy='most_frequent')
X_train_age=si_age.fit_transform(X_train[['Age']])
X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])

X_test_age=si_age.fit_transform(X_test[['Age']])
X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])
```

ohe one hot encoding on sex and Embarked

```python
ohe_sex=OneHotEncoder(sparse=False,handle_unknown='ignore')
ohe_embarked=OneHotEncoder(sparse=False,handle_unknown='ignore')
X_train_sex=ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked=ohe_embarked.fit_transform(X_train_embarked)

X_test_sex=ohe_sex.fit_transform(X_test[['Sex']])
X_test_embarked=ohe_embarked.transform(X_test_embarked)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe_sex = OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Fit and transform the training data
X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train[['Embarked']])

# Transform the test data
X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test[['Embarked']])
```

```python
X_train_rem=X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem=X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed=np.concatenate((X_train_rem,X_train_sex,X_train_embarked),axis=1)
X_test_transformed=np.concatenate((X_test_rem,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
X_train_transformed.shape
```

```python
X_test_transformed.shape
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
DecisionTreeClassifier()
```

```python
y_pred=clf.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
